```python
#creating the plot and adding a label for the legend.
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
#creating labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set y limit beteween 0 and 45*.
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare by Month")
#adding the legend.
plt.legend()
```

```python
#creating the plot and adding a label for the legend.
# Create the plot.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
#setting Boston Label*. 
ax.plot(x_axis, y_axis, marker="*", color="Green", linewidth=2, label='Boston')
#creating labels for the x and y axes
ax.set_ylabel("Fare($)")
ax.set_xlabel("Date")
#sting y limit between 0 and 45*.
ax.set_ylim(0,45)
#create a titile*.
ax.set_title("PyBer Fare by Month")
#adding a grid*.
ax.grid()
#adding the legend.
ax.legend()
```
