```python
#Description of prior outputs*.
#the plt.subplot() function returns a tuple (fig,ax) that contains () the tuple elements 
#when written as fig,ax=plt.subplots(), it unpacks the tuple into the variables fig and ax.
#the first variable, herein labeled as 'fig' references a figure
#the second variable, herein labeled as 'ax' references the axes of the figure 
#(the item listed before it) and allows annotation of hte figure
#ax.plot(x,y) plots the data from the x and y axes against the axes.
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#When we use the plt.plot() function to create a scatter plot, 
#we need to add a lowercase "o" as a parameter inside the parentheses. 
#This switches the plot from a line chart to a scatter plot.
plt.plot(x_axis,y_axis,'o')

```

```python
#To create a scatter plot using the plt.scatter() function, 
#add the x-axis and y-axis parameters inside the parentheses, 
#as shown here:
plt.scatter(x_axis, y_axis)

```

```python
#testing example:
plt.scatter(y_axis,x_axis,
         color='red',
         label='Chicago',
         marker='*')
plt.xlabel("Fare")
plt.ylabel('Date')
plt.xlim(0,50)
plt.title("PyBer Fare By Month")
plt.gca().invert_yaxis()
plt.legend()
```

```python
#the 's' parameter is the size parameter for the markers;
#below, we set the value of hte size of a given mark to its 
#value along the y_axis value*.
#(IE: size of markers will vary depending on value associated with y)

plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#in the below sequence, we multiple the size of the data in the yaxis
#by 3*. 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

#this is effectively creating a new variable for y_axis_larger 
#for each position of the data along the y values, we multiply by 3
#and add it to the new variable (list) we then set these values for 's'*.
plt.scatter(x_axis, y_axis, s=y_axis_larger)

```

```python
#You can use list comprehension to replace many for and while loops. 
#List comprehension is faster because it is optimized for Python 
#to spot a predictable pattern during looping.

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])

#this method is the same as the one wherein you created the loop 
#sequence; however, it uses a list comprehsnions method instead*.
```

```python
#below is the object oriented method instead*. 
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#testing with object oriented method of creation*.
#NOTE: ax.set_ylabel - you must include the subcommand set_ylabel*.
#this is different from the plt method of .ylabel*.
#NOTE: do not include = sign in the statements*.

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=300, color='skyblue', edgecolors='k',
           linewidth=2, label="Boston")
ax.set_title("Pyber Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare")
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.legend()
```

```python
#Just like plt.barh() creats a horizontal bar chart, plt.pie()
#makes a pie chart*.
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#pie chart explosions are paramters in the plt.pie functions*.
#NOTE: the explosion values variable can be a list [ ] or a tuple ()
#NOTE: length of the sequence should be equal to number of pie wedges*.
#NOTE: teh value speciifed to explode is the value that coincides with 
#the same index value in the plotted values against the pie explode values
#NOTE: this is an alternate method of writing out color values*.
#first, create a list with the total number of values matching the values 
#in the chart; then identify colords as the name assigned to the list*.

#creating color list*.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", 
          "greenyellow", "yellow", "orange", "gold", "indianred", 
          "tomato", "mistyrose"]
#executing pie chart graph*
plt.subplots(figsize=(8,8))
explode_values = [0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
plt.pie(y_axis, explode=explode_values,
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')
        
```

```python
#below is the object oriented method of pie charts*. 
#note: figure size changes is a function of the plt.subplots*.
#you cannot add figures size changes to the ax.pie methods*
#the figure size is set before the execution of the graphic*
#you must set figure size prior to identifying content of the graphic*. 
ax_explode = [0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis,
       colors=colors,
       shadow=True,
       autopct='%.1f%%',
       rotatelabels=False,
       explode=ax_explode,
       startangle=90)
ax.invert_xaxis()
plt.show()
```

```python
#subsequent sections will use the matplot lib
#to execute pyber manipulations*. 

# Import dependencies.
%matplotlib inline
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
#here we plot x and y; yerr=stdev plots values of error terms 
#capsize creates a cap on the graph end points for enhanced visuals*.
#the 'yerr' is the error values for the y axis;*
#to use error values on x axis, use xerr*.
plt.errorbar(x_axis, y_axis, yerr=stdev,capsize=3)
```

```python
#below is the object oriented approach*. 

fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#here you use the plt.bar functino to execute the error graphcs
#on a barchart*.

plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#below is object oriented approach*.
fig, ax=plt.subplots()
ax.bar(x_axis,y_axis,yerr=stdev,capsize=3)
plt.show()
```

```python
#major ticks are different from line ticks*
#to modify line ticks use the plt.subplt() function*
#to modfiy major ticks, do the following:
#this sets the increments in the ticks from 0 to 50 and sets 
#increments at every 5 values*.
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
#numpy is a numerical maths library that can be used to creaet
#arrays or matrices of numbers*. 
#below is  object oriented version*.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
#below is additional method of adding minor ticks to a graph*.
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```

```python
#5.1.10 - plotting panda frames*. 
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
#here you declare the x and y values according to their labels within
#the csv file / panda data array/ frame*.
#the label of the graphs values are already complete; they are 
#derived from values stored in the csv files*. 
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.

#Here you create an array using numpy arrange function 
#the arry is specified as containing the index values for the length
#of values contained in the uberride_df data set*. 

#then you send tick locations as being every value (month) in the xaxis
#using list comprehension*. 
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
#executing the same function using different procedures*.
#the depictino of the error values is contained within teh graphic 
#chart visuals*. 
#use the rotation method of plt.xticks(rotation) to do a rotate*.

pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)",
                       color="blue",
                       yerr=stdev, capsize=3)
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.yticks(np.arange(0, 51, step=5.0))
plt.xticks(rotation=0)
plt.show()

```
